# Rabin Games

The module `RabinGames` defines two classes:

* Game - represents a Rabin Game
* Node - represents nodes in the game

## The textual format

A `Game` is created from a string. The string can optionally have two parts: `nodes` and `dot_attr` (optinional) separated by `--DOT` line

    nodes
    --DOT--
    dot_attr
    
The part `nodes` contains description of the nodes, separated by a newline. A node is encoded by:

    [id] * green * red * succ
    (id) * green * red * succ
where:
* `id` is an integer,
* `green`, `red`, and `succ` are list of integers devided by a blank space ` `,
* `green` and `red` specifies a set of green and red sets the node belongs to,
* `succ` is a list of ids of successor nodes,
* `(id)` are nodes of the player _SYSTEM_,
* `[id]` are nodes of the player _ENVIRONMENT_.

The `dot_attr` part is included at the beginning of the graph in dot representation. See the Section [#Visualisation](#Visualisation) for more details.

```python
test = '''
(1) * 1 2 * 2 * 2 3
[2] * * 1 * 1
[3] * 2 * * 1
'''
```

The string `test` describes a game of three nodes, environment has two nodes, the player system has one. You can create game simply using strings:

```python
Game(test)
```

or from file:

```python
print (open('ce1.game','r').read())
game = Game(filename='ce1.game')
```

## Visualisation

Both classes `Game` and `Node` has a method `get_dot()` which prints their representation in the graphiz `dot` language. The method `Node.get_dot()` does not produce a whole dot-graph, it is only used by the `Game.get_dot()` method. The class `Game` has two methods for obtaining SVG:

* `Game.svg_to_file(filename=game.svg)` a method that creates an `.svg` file `filename`, and
* `Game.to_svg()` which returns an SVG source

In IPython you can use commands

* `game` as a last command of a cell (goes to the output), or
* `IPython.display.display(game)` anywhere in the cell (like print)

to visualize the object `game` as an SVG image.

```python
testgame = Game(test)
IPython.display.display(testgame,game) # Let's show two games by one command. Yeah!
testgame # And show the image of the test game in the ouput of the cell
```

As you can see, the `game` is rendered nicely. This is done by the `dot_attr` part of the game textual representation. Let's see what does it look like if no formatting is used. And let's see how can I easily change the direction of the graph.

```python
game.dot_attr = ''
print 
IPython.display.display(game,'PHH! Takes too much space')
game.dot_attr = 'rankdir=LR\n'
game
```

Let's make it a circle!

```python
game.dot_attr += '{rank=same 2 4}'
IPython.display.display(game,'Not very good. We still need to put 1 back to the beginning.')
game.dot_attr = '1\n' + game.dot_attr
game
```

### dot_attr

#### rankdir

There are four directions:

* LR
* TB
* RL
* BT

The line `{rank=same 2 1}` defines a subgraph of the nodes 2 and 1 and tells dot to draw them at one level.

```python
Game(test,dot_attr='{rank=same 2 1}\n')
```

You can of course use any of the global dot attributes. The node attributes `label` and `shape` are overwritten, others should work fine. Formatting an edge will draw a new edge.

```python
Game(test,dot_attr='rankdir=LR\n2 [color=blue,label="Hello"]\n1 -> 3 [color=red]\n2 [shape=circle]')
```
